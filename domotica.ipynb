```python
#Heater

temp_actual = float(input("temperatura actual: "))

class calefactor:
    def __init__(self, limit):
        self.limit = limit
        self.ences = False

    def actualizar(self, temp_actual):
        if temp_actual < self.limit:
            self.ences = True
            print("Calefactor ences")
        else:
            self.ences = False
            print("Calefactor apagat")


opcion = ""
calefactor_inst = None  

while opcion != "c":
    print("Menu:")
    print("a) configuar temperatura")
    print("b) ver estado")
    print("c) salir")
    
    opcion = input("Elige una opcion: ").lower().strip()

    if opcion == "a":
        limit = float(input("temperatura limit: "))
        calefactor_inst = calefactor(limit)
        calefactor_inst.actualizar(temp_actual)

    elif opcion == "b":
        if calefactor_inst is None:
            print("configura la temperatura abans")
        else:
            calefactor_inst.actualizar(temp_actual)

    elif opcion == "c":
        print("adeu")

    else:
        print("x")
```

```python
#Automatic sprinkler

class Aspersor:
    def __init__(self):
        self.hora_activacion = None
        self.duracion = None
        self.estado = "Inactivo"

    def configurar(self):
        self.hora_activacion = input("Hora de activacion (HH:MM): ")
        self.duracion = float(input("Duracion (min): "))
        print("OK")

    def ver_estado(self):
        print(f"Duracion: {self.duracion} min")
        print(f"Estado: {self.estado}")

    def cancelar(self):
        self.hora_activacion = None
        self.duracion = None
        self.estado = "Inactivo"
        print("Cancelado")

def main():
    asp = Aspersor()
    while True:
        print("Menú principal:")
        print("[a] Configurar riego")
        print("[b] Ver estado actual")
        print("[c] Cancelar riego")
        print("[d] Salir")
        op = input("Opción: ")
        if op == "a":
            asp.configurar()
        elif op == "b":
            asp.ver_estado()
        elif op == "c":
            asp.cancelar()
        elif op == "d":
            break
        else:
            print("De la a-d")

if __name__ == "__main__":
    main()
```
