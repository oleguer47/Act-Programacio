```python
import datetime

nombre = ""
edad = 0
peso = 0.0
altura = 0.0

opcion = ""
while opcion != "d":
    print("Menu:")
    print("a) Introducir datos")
    print("b) Modificar datos")
    print("c) Ver datos")
    print("d) Salir")

    opcion = input("Elige una opción: ").lower().strip()

    if opcion == "a":
       
        nombre = input("Nombre completo: ").strip()
        if nombre == "":
            print("No puede estar vacio")
            continue
        nombre = nombre[0].upper() + nombre[1:]

        valor = input("Edad: ")
        try:
            edad = int(valor)
            if edad <= 0 or edad > 120:
                print("Edad invalida")
                continue
        except:
            print("Escribe un numero entero")
            continue

        valor = input("Peso (kg): ").replace(",", ".")
        try:
            peso = float(valor)
            if peso <= 0 or peso > 400:
                print("Peso invalido")
                continue
        except:
            print("Escribe un numero")
            continue

        valor = input("Altura (m): ").replace(",", ".")
        try:
            altura = float(valor)
            if altura <= 0.5 or altura > 2.5:
                print("Altura invalida")
                continue
        except:
            print("Escribe un numero")
            continue

        print("Datos guardados correctamente.")

    elif opcion == "b":
        
        if nombre == "":
            print("Primero introduce los datos.")
            continue
        campo = input("¿Qué quieres modificar (nombre, edad, peso, altura)? ").lower().strip()
        if campo == "nombre":
            nuevo = input("Nuevo nombre: ").strip()
            if nuevo:
                nombre = nuevo
        elif campo == "edad":
            valor = input("Nueva edad: ")
            try:
                nueva = int(valor)
                if nueva > 0 and nueva <= 120:
                    edad = nueva
                else:
                    print("Edad invalida")
            except:
                print("Escribe un numero entero")
        elif campo == "peso":
            valor = input("Nuevo peso (kg): ").replace(",", ".")
            try:
                nuevo = float(valor)
                if nuevo > 0 and nuevo <= 400:
                    peso = nuevo
                else:
                    print("Peso invalido")
            except:
                print("Escribe un numero")
        elif campo == "altura":
            valor = input("Nueva altura (m): ").replace(",", ".")
            try:
                nuevo = float(valor)
                if nuevo > 0.5 and nuevo <= 2.5:
                    altura = nuevo
                else:
                    print("Altura invalida")
            except:
                print("Escribe un numero")
        else:
            print("Campo no válido.")

    elif opcion == "c":
        
        if not nombre:
            print("No hay datos.")
            continue

        imc = peso / (altura * altura)
        if imc < 18.5:
            categoria = "Peso bajo"
        elif imc < 25:
            categoria = "Peso normal"
        elif imc < 30:
            categoria = "Sobrepeso"
        else:
            categoria = "Obesidad"

        
        fc_max = 220 - edad
        fc_min = int(fc_max * 0.5)
        fc_max_zona = int(fc_max * 0.85)
        agua = peso * 0.035
        año_actual = datetime.date.today().year
        año_nacimiento = año_actual - edad

        print("Resultados")
        print(f"Hola, {nombre.title()}!")
        print(f"Edad: {edad} años | peso: {peso} kg | altura: {altura} m")
        print(f"IMC: {imc:.2f} ({categoria})")
        print(f"FC màxima estimada: {fc_max} bpm")
        print(f"Zona FC objectiu: {fc_min}–{fc_max_zona} bpm")
        print(f"Aigua recomanada: {agua:.2f} Litres al dia")
        print(f"Any de naixement aproximat: {año_nacimiento}\n")

    elif opcion == "d":
        print("Adeu")
    else:
        print("Opcio no valida.")
```
