```python
import time

class Calefactor:
    def __init__(self, limit):
        self.limit = limit
        self.encès = False

    def activar(self):
        self.encès = True
        print("Calefactor encès")

    def desactivar(self):
        self.encès = False
        print("Calefactor apagat")

print("calefactor:")

limit = float(input("Introdueix la temperatura limit (°C): "))
temp_actual = float(input("Introdueix la temperatura actual (°C): "))

calefactor = Calefactor(limit)

print("temperatura:")

if temp_actual < limit:
    calefactor.activar()
    print(f"Temperatura pujant fins arribar a {limit}°C")
    for t in range(int(temp_actual), int(limit) + 1):
        print(f"Temperatura: {t}°C")
        time.sleep(0.2)
    calefactor.desactivar()

elif temp_actual > limit:
    calefactor.desactivar()
    print(f"Temperatura baixant fins arribar a {limit}°C")
    for t in range(int(temp_actual), int(limit) - 1, -1):
        print(f"Temperatura: {t}°C")
        time.sleep(0.2)
    calefactor.activar()

else:
    print(f"La temperatura és igual a ({limit}°C). No s’activa el calefactor.")

print("OK")
```

```python
import time
from datetime import datetime

class Aspersor:
    def __init__(self):
        self.hora_activacion = None
        self.duracion = None
        self.estado = "Inactivo"

    def configurar(self):
        self.hora_activacion = input("Hora de activacion (HH:MM): ")
        self.duracion = float(input("Duracion (min): "))
        print("Configurado correctament. Esperant activació...")
        self.simular()

    def simular(self):
        hora_actual = datetime.now().strftime("%H:%M")
        while hora_actual != self.hora_activacion:
            print(f"Hora actual: {hora_actual} — Aspersor {self.estado}")
            time.sleep(1)
            hora_actual = datetime.now().strftime("%H:%M")
        self.activar()

    def activar(self):
        self.estado = "Activat"
        print(f"Aspersor activat a les {self.hora_activacion}")
        for i in range(int(self.duracion)):
            print(f"Minut {i+1}/{int(self.duracion)} de reg")
            time.sleep(1)
        self.desactivar()

    def desactivar(self):
        self.estado = "Inactiu"
        print("Aspersor desactivat")

asp = Aspersor()
asp.configurar()
```
